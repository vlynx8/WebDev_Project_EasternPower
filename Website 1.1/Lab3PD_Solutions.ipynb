## Range
Write a Python program that reads in a number and checks if it is between 1 and 100 (inclusive).

```python
a = float(input("Please enter a positive integer:\n"))

if(a >=1 and a<=100):
    print('in range')
else:
    print('out of range')
```

Can you do it using 'or' without 'and'?

```python
a = float(input("Please enter a positive integer:\n"))
if not(a <1 or a>100):
    print('in range')
else:
    print('out of range')
```

## Print the first n odd numbers using a while loop.
Write a Python program to print the first n (a positive integer read in from the user) odd numbers using a while loop. 

```python
a = int(input("Please enter a positive integer:\n"))
i=0
odd = 1
while i<a:
    i = i+1
    print(odd)
    odd = odd + 2
```

## Print numbers from n down to 1 using a while loop.
Write a Python program to print numbers from some integer, n, down to 1 using a while loop.

```python
a = int(input("Please enter a positive integer:\n"))
while a>0:
    print(a)
    a = a-1
```

## Grade a student based on their marks
Write code that, upon entering a mark (type float), prints a students grade with the following specifications.

- A - 85% or over
- B - 70% or over
- C - 60% or over
- D - 40% or over
- F - less than 40%

Can you do this using only one greater than (>) or less than (<) comparison per grade?


```python
result = float(input("Please enter a positive integer:\n"))
if (result<40):
    print('F')
elif (result<60):
    print('D')
elif (result<70):
    print('C')
elif (result<85):
    print('B')
else:
    print('A')
```
